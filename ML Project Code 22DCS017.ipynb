```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
import wandb

from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, KFold
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE 

# --- NEW IMPORTS FOR USER REQUEST ---
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture  # For Latent Space / Model-based clustering
from sklearn.naive_bayes import GaussianNB   # For Naive Bayes Validation

from scipy.stats import trim_mean, tstd
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score, accuracy_score

warnings.filterwarnings('ignore')

print("All libraries imported!")
# Note: Run wandb.login() in a separate cell before this.

# --- Helper Functions ---

def calculate_vc_dimension(n_samples, n_features, n_clusters, model_type="KMeans"):
    """
    Estimates the VC-Dimension (Vapnik-Chervonenkis) complexity.
    Theoretical bound for K-Means is roughly O(K * d).
    """
    if model_type == "KMeans" or model_type == "GMM":
        # Approx VC dim: K clusters * (Dimensions + 1)
        vc_dim = n_clusters * (n_features + 1)
        return vc_dim
    return "Undefined for this algo"

def run_experiments(X_features, optimal_k, dataset_name):
    print(f"\n--- Running Experimental Setups (Train/Test Splits) for {dataset_name} ---")
    
    distributions = {'50-50': 0.5, '60-40': 0.4, '70-30': 0.3, '80-20': 0.2}
    
    # Added GaussianMixture (Latent Space Method) here
    models_to_test = {
        'KMeans': KMeans(n_clusters=optimal_k, init='k-means++', n_init=10, random_state=42),
        'Agglomerative': AgglomerativeClustering(n_clusters=optimal_k),
        'GMM (Latent Space)': GaussianMixture(n_components=optimal_k, covariance_type='full', random_state=42)
    }
    
    for model_name, model_instance in models_to_test.items():
        print(f"\n... Testing Model: {model_name} ...")
        model_results = {}
        
        for name, test_size in distributions.items():
            # Ensure X_features is not empty
            if len(X_features) == 0:
                model_results[name] = "Error - No Data"
                continue
                
            # Ensure test_size doesn't result in an empty train set
            if int(len(X_features) * (1 - test_size)) < 1:
                 model_results[name] = "Error - Invalid Split"
                 continue
                 
            # Ensure n_clusters is not greater than n_samples
            if optimal_k > int(len(X_features) * (1 - test_size)):
                model_results[name] = "Error - n_clusters > n_samples"
                continue

            X_train, X_test = train_test_split(X_features, test_size=test_size, random_state=42)
            
            # Check again for train and test sets
            if X_train.shape[0] < optimal_k or X_test.shape[0] < optimal_k:
                model_results[name] = "Error - Not enough samples for K"
                continue

            model = model_instance
            
            # Fit and Predict logic handling for different models
            if "Agglomerative" in model_name:
                # Agglomerative doesn't have a predict method for new data in sklearn standard, 
                # so we fit_predict separately for train and test to get metrics (approximate evaluation)
                model.fit(X_train)
                train_labels = model.labels_
                test_model = AgglomerativeClustering(n_clusters=optimal_k)
                test_labels = test_model.fit_predict(X_test)
            else:
                model.fit(X_train)
                if "GMM" in model_name:
                    train_labels = model.predict(X_train)
                    test_labels = model.predict(X_test)
                else: # KMeans
                    train_labels = model.labels_
                    test_labels = model.predict(X_test)

            # --- VC Dimension Calculation ---
            n_feats = X_train.shape[1]
            vc_dim = calculate_vc_dimension(len(X_train), n_feats, optimal_k, model_type="GMM" if "GMM" in model_name else "KMeans")
            
            # --- Naive Bayes Validation (Supervised check on Unsupervised labels) ---
            # We train a NB classifier to see if it can learn the cluster boundaries easily
            nb_validator = GaussianNB()
            try:
                nb_validator.fit(X_train, train_labels)
                nb_pred = nb_validator.predict(X_test)
                nb_accuracy = accuracy_score(test_labels, nb_pred)
            except:
                nb_accuracy = np.nan

            try:
                model_results[name] = {
                    'Silhouette (Train)': silhouette_score(X_train, train_labels),
                    'Calinski (Train)': calinski_harabasz_score(X_train, train_labels),
                    'Davies-Bouldin (Train)': davies_bouldin_score(X_train, train_labels),
                    'Silhouette (Test)': silhouette_score(X_test, test_labels),
                    'Calinski (Test)': calinski_harabasz_score(X_test, test_labels),
                    'VC-Dimension (Approx)': vc_dim,
                    'NB-Validation Acc': nb_accuracy  # How well Naive Bayes predicts these clusters
                }
            except Exception as e:
                print(f"  Error calculating metrics for {name}: {e}")
                model_results[name] = f"Error: {e}"
                
        results_df = pd.DataFrame(model_results).T
        print(f"[Results for {model_name} - {dataset_name}]")
        print(results_df.round(3))
        
        results_df_for_log = results_df.reset_index().rename(columns={'index':'Split'})
        wandb.log({f"exp_setup_{model_name}": wandb.Table(dataframe=results_df_for_log)})

def run_kmeans_cv(X, n_clusters, n_splits=5, dataset_name=""):
    print(f"\n--- Running {n_splits}-Fold Cross-Validation for {dataset_name} (KMeans, K={n_clusters}) ---")
    
    if X.shape[0] < n_splits:
        print(f"  Skipping CV: Not enough samples ({X.shape[0]}) for {n_splits} splits.")
        return
        
    if X.shape[0] < n_clusters:
        print(f"  Skipping CV: Not enough samples ({X.shape[0]}) for {n_clusters} clusters.")
        return

    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    
    fold_metrics = {
        'Silhouette (Train)': [], 'Calinski (Train)': [], 'Davies-Bouldin (Train)': [],
        'Silhouette (Test)': [],  'Calinski (Test)': [],  'Davies-Bouldin (Test)': []
    }

    for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
        X_train_cv, X_test_cv = X[train_idx], X[test_idx]
        
        if X_train_cv.shape[0] < n_clusters or X_test_cv.shape[0] < n_clusters:
            print(f"  Skipping fold {fold+1}: Not enough samples in split for K={n_clusters}.")
            continue

        model = KMeans(n_clusters=n_clusters, init='k-means++', n_init=10, random_state=42)
        model.fit(X_train_cv)
        train_labels = model.labels_
        test_labels = model.predict(X_test_cv)
        
        try:
            fold_metrics['Silhouette (Train)'].append(silhouette_score(X_train_cv, train_labels))
            fold_metrics['Calinski (Train)'].append(calinski_harabasz_score(X_train_cv, train_labels))
            fold_metrics['Davies-Bouldin (Train)'].append(davies_bouldin_score(X_train_cv, train_labels))
            fold_metrics['Silhouette (Test)'].append(silhouette_score(X_test_cv, test_labels))
            fold_metrics['Calinski (Test)'].append(calinski_harabasz_score(X_test_cv, test_labels))
            fold_metrics['Davies-Bouldin (Test)'].append(davies_bouldin_score(X_test_cv, test_labels))
        except Exception as e:
            print(f"  Error in fold {fold+1} metrics: {e}")
            pass
            
    print(f"Cross-Validation Results (Average of {len(fold_metrics['Silhouette (Test)'])} successful folds):")
    avg_results = {k: np.mean(v) if v else np.nan for k, v in fold_metrics.items()}
    print(pd.Series(avg_results).round(3))
    
    cv_log_dict = {f"cv_KMeans_{k}": v for k, v in avg_results.items()}
    wandb.log(cv_log_dict)

def run_dbscan_analysis(X_features, eps, min_samples, dataset_name):
    print(f"\n... Testing Model: DBSCAN ...")
    if X_features.shape[0] < min_samples:
        print(f"  Skipping DBSCAN: Not enough samples ({X_features.shape[0]}) for min_samples={min_samples}.")
        return

    try:
        dbscan = DBSCAN(eps=eps, min_samples=min_samples)
        dbscan.fit(X_features)
        db_labels = dbscan.labels_
        n_clusters = len(set(db_labels)) - (1 if -1 in db_labels else 0)
        print(f"DBSCAN found {n_clusters} clusters (plus noise).")

        if n_clusters > 1 and n_clusters < (len(X_features) - 1):
            dbscan_results = {
                'Silhouette': silhouette_score(X_features, db_labels),
                'Calinski': calinski_harabasz_score(X_features, db_labels),
                'Davies-Bouldin': davies_bouldin_score(X_features, db_labels)
            }
            print(f"[Results for DBSCAN (Full Data) - {dataset_name}]")
            print(pd.Series(dbscan_results).round(3))
            
            dbscan_log_dict = {f"dbscan_{k}": v for k, v in dbscan_results.items()}
            dbscan_log_dict["dbscan_clusters_found"] = n_clusters
            wandb.log(dbscan_log_dict)
        else:
            print("DBSCAN found 0 or 1 cluster. Metrics cannot be calculated.")
            wandb.log({"dbscan_clusters_found": n_clusters})
            
    except Exception as e:
        print(f"  Error running DBSCAN: {e}")


# --- DATASET 1: Ad Events ---

print("\n\n" + "="*80)
print("STARTING DATASET 1: Ad Events")
print("="*80)

ds1_run = wandb.init(
    project="social-media-segmentation",
    name="Dataset 1 - Ad Events",
    reinit=True
)

try:
    print("Step 1: Loading data...")
    # Make sure to update these paths to your local file locations
    users_df = pd.read_csv('D:\\Project\\Dataset 1\\users.csv')
    ad_events_df = pd.read_csv('D:\\Project\\Dataset 1\\ad_events.csv')
    print("Dataset 1 files loaded.")

    print("\nStep 2: Combining and engineering features...")
    behavioral_df = ad_events_df.groupby('user_id')['event_type'].value_counts().unstack(fill_value=0)
    behavioral_df.columns = [f'num_{col.lower()}s' for col in behavioral_df.columns]
    behavioral_df = behavioral_df.reset_index()
    behavioral_df['total_events'] = behavioral_df.drop('user_id', axis=1).sum(axis=1)
    behavioral_df['ctr'] = (behavioral_df['num_clicks'] / behavioral_df['num_impressions'])
    behavioral_df['conversion_rate'] = (behavioral_df['num_purchases'] / behavioral_df['num_clicks'])
    behavioral_df.replace([np.inf, -np.inf], 0, inplace=True)
    behavioral_df.fillna(0, inplace=True)
    df_profiled = pd.merge(users_df, behavioral_df, on='user_id', how='left')
    behavioral_cols = behavioral_df.drop('user_id', axis=1).columns
    df_profiled[behavioral_cols] = df_profiled[behavioral_cols].fillna(0)
    print("Feature engineering complete.")

    print("\nStep 3: Analyzing & Cleaning data...")
    
    # Analyze missing age data *before* imputation
    initial_rows = len(df_profiled)
    missing_age = df_profiled['user_age'].isna().sum()
    print(f"Found {missing_age} rows ({missing_age/initial_rows*100:.1f}%) with missing 'user_age'.")
    
    # Drop rows where 'user_age' is missing, as it's a key profiling feature.
    df_profiled = df_profiled.dropna(subset=['user_age'])
    rows_after_age_drop = len(df_profiled)
    print(f"Removed {initial_rows - rows_after_age_drop} rows due to missing age. Remaining rows: {rows_after_age_drop}")
    
    if df_profiled.empty:
        raise Exception("No data remaining after dropping rows with missing age. Stopping execution for Dataset 1.")

    print("\nAnalyzing outliers...")
    plt.figure(figsize=(15, 5))
    sns.boxplot(data=df_profiled[['total_events', 'num_clicks', 'num_purchases']], orient='h')
    plt.title('Boxplots for Dataset 1 (Before Cleaning)')
    plt.savefig('ds1_boxplot.png') 
    wandb.log({"Data Analysis/Boxplot (Before)": wandb.Image('ds1_boxplot.png')}) 
    plt.show()
    
    Q1 = df_profiled['total_events'].quantile(0.25)
    Q3 = df_profiled['total_events'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df_profiled[(df_profiled['total_events'] >= lower_bound) & (df_profiled['total_events'] <= upper_bound)].copy()
    rows_removed = len(df_profiled) - len(df_cleaned)
    print(f"Removed {rows_removed} outliers based on 'total_events'.")
    
    if df_cleaned.empty:
        raise Exception("No data remaining after outlier removal. Stopping execution for Dataset 1.")

    print(f"\n--- Data Preview (DS1 After Trimming) ---")
    print(df_cleaned.head())
    print("\nCleaned Data Description (DS1):")
    print(df_cleaned.describe().round(2))

    stats = {
        "mean": df_cleaned['total_events'].mean(),
        "median": df_cleaned['total_events'].median(),
        "q1": df_cleaned['total_events'].quantile(0.25),
        "q3": df_cleaned['total_events'].quantile(0.75),
        "trimmed_mean_10": trim_mean(df_cleaned['total_events'], 0.1),
        "trimmed_std_dev": tstd(df_cleaned['total_events'], limits=(Q1, Q3))
    }
    print(f"\n[Descriptive Stats for 'total_events' (Cleaned)]")
    for k, v in stats.items(): print(f"  {k}: {v:.2f}")
    wandb.log({"Data Analysis/Stats": stats})
    wandb.log({"Data Analysis/Rows Removed (Missing Age)": initial_rows - rows_after_age_drop})
    wandb.log({"Data Analysis/Rows Removed (Outlier)": rows_removed})
    
    plt.figure(figsize=(10, 6))
    numeric_df = df_cleaned.select_dtypes(include=np.number)
    if 'user_id' in numeric_df.columns: numeric_df = numeric_df.drop('user_id', axis=1)
    sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.1f')
    plt.title('Correlation Heatmap (Dataset 1)')
    plt.savefig('ds1_corr.png') 
    wandb.log({"Data Analysis/Correlation": wandb.Image('ds1_corr.png')}) 
    plt.show()

    print("\nStep 4: ML Pipelines...")
    numerical_features = ['user_age', 'num_clicks', 'num_comments', 'num_impressions', 'num_likes', 'num_purchases', 'num_shares', 'total_events', 'ctr', 'conversion_rate']
    categorical_features = ['user_gender']
    text_features = 'interests'
    
    # Imputer is still good practice in case other num features have NaNs
    preprocessor_ds1 = ColumnTransformer(
        transformers=[
            ('num', Pipeline([('imputer', SimpleImputer(strategy='mean')), ('scaler', StandardScaler())]), numerical_features),
            ('cat', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))]), categorical_features),
            ('text', Pipeline([('tfidf', TfidfVectorizer(max_features=50))]), text_features)
        ], remainder='drop')
        
    X_features_ds1 = preprocessor_ds1.fit_transform(df_cleaned)
    if hasattr(X_features_ds1, 'toarray'):
        X_features_ds1 = X_features_ds1.toarray()
    print(f"Features created. Shape: {X_features_ds1.shape}")

    print("\nStep 5: Finding optimal K (Elbow Method)...")
    wcss = []
    # Ensure K is not larger than the number of samples
    max_k = min(10, X_features_ds1.shape[0] - 1)
    if max_k < 2:
        raise Exception(f"Not enough samples ({X_features_ds1.shape[0]}) to run KMeans.")
        
    k_range = range(2, max_k + 1)
    print(f"Running Elbow method for K from 2 to {max_k}")
    
    for k in k_range:
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
        kmeans.fit(X_features_ds1)
        wcss.append(kmeans.inertia_)
        wandb.log({"Elbow/WCSS": kmeans.inertia_, "Elbow/K": k}) 
    
    plt.figure(figsize=(8, 4))
    plt.plot(k_range, wcss, 'o--')
    plt.title('Elbow Method (Dataset 1)')
    plt.xlabel('K Clusters'); plt.ylabel('WCSS')
    plt.savefig('ds1_elbow.png') 
    wandb.log({"Elbow/Elbow Plot": wandb.Image('ds1_elbow.png')}) 
    plt.show()
    
    # Choose K. Default to 4 if possible, else adjust.
    OPTIMAL_K_DS1 = 4 if max_k >= 4 else max_k
    print(f"Chosen K for Dataset 1: {OPTIMAL_K_DS1}")
    wandb.config.update({"optimal_k": OPTIMAL_K_DS1}) 

    run_experiments(X_features_ds1, OPTIMAL_K_DS1, "Dataset 1")
    run_kmeans_cv(X_features_ds1, OPTIMAL_K_DS1, n_splits=5, dataset_name="Dataset 1")
    run_dbscan_analysis(X_features_ds1, eps=2.5, min_samples=10, dataset_name="Dataset 1")

    print("\nStep 7: Final Visualization & Profiling...")
    final_model = KMeans(n_clusters=OPTIMAL_K_DS1, init='k-means++', n_init=10, random_state=42)
    labels = final_model.fit_predict(X_features_ds1)
    df_cleaned['segment_id'] = labels
    
    # --- SEGMENT PROFILING & NAMING (DS1) ---
    print("\n--- Segment Profiles (Dataset 1) ---")
    numerical_profile = df_cleaned.groupby('segment_id')[numerical_features].mean()
    categorical_profile = df_cleaned.groupby('segment_id')[categorical_features].agg(lambda x: x.mode().iloc[0])
    
    try:
        # Get the feature names *and* the number of text features
        text_feature_names = preprocessor_ds1.named_transformers_['text'].named_steps['tfidf'].get_feature_names_out()
        n_text_features = len(text_feature_names)
        
        # Slice only the last 'n_text_features' from the cluster centers
        cluster_centers_tfidf = final_model.cluster_centers_[:, -n_text_features:]
        
        top_interests = []
        for i in range(OPTIMAL_K_DS1):
            # Indices will now correctly be from 0 to (n_text_features - 1)
            top_indices = cluster_centers_tfidf[i].argsort()[-3:][::-1]
            top_interests.append(', '.join([text_feature_names[idx] for idx in top_indices]))
        categorical_profile['Top_Interests'] = top_interests
        
    except Exception as e:
        print(f"Could not get interests for profiling: {e}")
    
    print("\n[Numerical Profile (Mean Values)]")
    print(numerical_profile.round(2))
    wandb.log({"Final/Numerical Profile": wandb.Table(dataframe=numerical_profile.reset_index())})
    
    print("\n[Categorical Profile (Mode & Top Interests)]")
    print(categorical_profile)
    wandb.log({"Final/Categorical Profile": wandb.Table(dataframe=categorical_profile.reset_index())})

    # --- DYNAMIC SEGMENT NAMING LOGIC (DS1) ---
    avg_age = df_cleaned['user_age'].mean()
    avg_ctr = df_cleaned['ctr'].mean()
    avg_conv_rate = df_cleaned['conversion_rate'].mean()

    def name_segment_ds1(row):
        name = []
        # Check Age
        if row['user_age'] > avg_age * 1.1: name.append("Older")
        elif row['user_age'] < avg_age * 0.9: name.append("Younger")
        
        # Check Conversion Rate
        if row['conversion_rate'] > avg_conv_rate * 1.5 and avg_conv_rate > 0: name.append("High Converters")
        elif row['conversion_rate'] < avg_conv_rate * 0.5 and avg_conv_rate > 0: name.append("Low Converters")

        # Check CTR
        if row['ctr'] > avg_ctr * 1.5 and avg_ctr > 0: name.append("High CTR")
        elif row['ctr'] < avg_ctr * 0.5 and avg_ctr > 0: name.append("Low CTR")
        
        if not name:
            return "General Average Users"
        return ", ".join(name)
    # --- END LOGIC ---

    profile_summary = numerical_profile.join(categorical_profile, how="left")
    segment_name_map = {idx: name_segment_ds1(row) for idx, row in profile_summary.iterrows()}
    
    print("\n--- Meaningful Segment Definitions (DS1) ---")
    print("Significance: These names describe the 'persona' of each cluster.")
    for seg_id, name in segment_name_map.items():
        print(f"Segment {seg_id}: {name}")
        
    df_cleaned['segment'] = df_cleaned['segment_id'].map(segment_name_map)
    # --- END PROFILING ---

    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_features_ds1)
    
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df_cleaned['segment'], alpha=0.6)
    plt.title('User Segments (PCA) - Dataset 1')
    plt.xlabel('PC1'); plt.ylabel('PC2')
    plt.savefig('ds1_pca.png') 
    wandb.log({"Final/PCA_Plot": wandb.Image('ds1_pca.png')}) 
    plt.show()
    
    df_cleaned.to_csv('Dataset1_Segmented.csv', index=False)
    print(">> Dataset 1 saved as 'Dataset1_Segmented.csv'")
    
except Exception as e:
    print(f"\n!!! Error in Dataset 1: {e} !!!")
finally:
    if ds1_run:
        wandb.finish() 


# --- DATASET 2: Social Media Engagement ---

print("\n\n" + "="*80)
print("STARTING DATASET 2: Social Media Engagement")
print("="*80)

ds2_run = wandb.init(
    project="social-media-segmentation",
    name="Dataset 2 - Engagement",
    reinit=True
)

try:
    print("Step 1 & 2: Loading and preparing data...")
    # Make sure to update these paths to your local file locations
    df_ds2 = pd.read_csv('D:\\Project\\Dataset 2\\Social Media Engagement Dataset.csv')
    df_ds2 = df_ds2.rename(columns={'engagement_rate': 'avg_engagement_rate'})
    print("Data loaded.")

    print("\nStep 3: Analyzing & Cleaning...")
    plt.figure(figsize=(15, 5))
    sns.boxplot(data=df_ds2[['avg_engagement_rate', 'likes_count', 'impressions']], orient='h')
    plt.title('Boxplots for Dataset 2 (Before Cleaning)')
    plt.savefig('ds2_boxplot.png') 
    wandb.log({"Data Analysis/Boxplot (Before)": wandb.Image('ds2_boxplot.png')}) 
    plt.show()

    Q1 = df_ds2['avg_engagement_rate'].quantile(0.25)
    Q3 = df_ds2['avg_engagement_rate'].quantile(0.75)
    IQR = Q3 - Q1
    df_cleaned_ds2 = df_ds2[df_ds2['avg_engagement_rate'] <= (Q3 + 1.5 * IQR)].copy()
    rows_removed = len(df_ds2) - len(df_cleaned_ds2)
    print(f"Removed {rows_removed} outliers.")
    wandb.log({"Data Analysis/Rows Removed": rows_removed}) 
    
    if df_cleaned_ds2.empty:
        raise Exception("No data remaining after outlier removal. Stopping execution for Dataset 2.")

    print(f"\n--- Data Preview (DS2 After Trimming) ---")
    print(df_cleaned_ds2.head())
    print("\nCleaned Data Description (DS2):")
    print(df_cleaned_ds2.describe().round(2))

    stats_ds2 = {
        "mean": df_cleaned_ds2['avg_engagement_rate'].mean(),
        "trimmed_mean_10": trim_mean(df_cleaned_ds2['avg_engagement_rate'], 0.1)
    }
    print(f"[Stats 'avg_engagement_rate'] Mean: {stats_ds2['mean']:.4f}")
    print(f"Trimmed Mean (10%): {stats_ds2['trimmed_mean_10']:.4f}")
    wandb.log({"Data Analysis/Stats": stats_ds2}) 

    plt.figure(figsize=(10, 6))
    sns.heatmap(df_cleaned_ds2.select_dtypes(include=np.number).drop(['post_id', 'user_id'], axis=1, errors='ignore').corr(), annot=True, cmap='coolwarm', fmt='.1f')
    plt.title('Correlation Heatmap (Dataset 2)')
    plt.savefig('ds2_corr.png') 
    wandb.log({"Data Analysis/Correlation": wandb.Image('ds2_corr.png')}) 
    plt.show()

    print("\nStep 4: ML Pipelines...")
    numerical_features_ds2 = ['sentiment_score', 'toxicity_score', 'avg_engagement_rate', 'likes_count', 'shares_count', 'comments_count', 'impressions', 'user_past_sentiment_avg', 'user_engagement_growth']
    categorical_features_ds2 = ['platform', 'topic_category', 'emotion_type']
    
    preprocessor_ds2 = ColumnTransformer(
        transformers=[
            ('num', Pipeline([('imputer', SimpleImputer(strategy='mean')), ('scaler', StandardScaler())]), 
             numerical_features_ds2),
            ('cat', Pipeline([('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))]), 
             categorical_features_ds2)
        ], remainder='drop')

    X_features_ds2 = preprocessor_ds2.fit_transform(df_cleaned_ds2)
    print(f"Features created. Shape: {X_features_ds2.shape}")

    print("\nStep 5: Finding optimal K...")
    wcss = []
    max_k_ds2 = min(10, X_features_ds2.shape[0] - 1)
    if max_k_ds2 < 2:
        raise Exception(f"Not enough samples ({X_features_ds2.shape[0]}) to run KMeans.")
    k_range_ds2 = range(2, max_k_ds2 + 1)
    print(f"Running Elbow method for K from 2 to {max_k_ds2}")

    for k in k_range_ds2:
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
        kmeans.fit(X_features_ds2)
        wcss.append(kmeans.inertia_)
        wandb.log({"Elbow/WCSS": kmeans.inertia_, "Elbow/K": k}) 
    
    plt.figure(figsize=(8, 4))
    plt.plot(k_range_ds2, wcss, 'o--')
    plt.title('Elbow Method (Dataset 2)')
    plt.savefig('ds2_elbow.png') 
    wandb.log({"Elbow/Elbow Plot": wandb.Image('ds2_elbow.png')}) 
    plt.show()
    
    OPTIMAL_K_DS2 = 4 if max_k_ds2 >= 4 else max_k_ds2
    print(f"Chosen K for Dataset 2: {OPTIMAL_K_DS2}")
    wandb.config.update({"optimal_k": OPTIMAL_K_DS2}) 

    run_experiments(X_features_ds2, OPTIMAL_K_DS2, "Dataset 2")
    run_kmeans_cv(X_features_ds2, OPTIMAL_K_DS2, n_splits=5, dataset_name="Dataset 2")
    run_dbscan_analysis(X_features_ds2, eps=1.5, min_samples=5, dataset_name="Dataset 2")

    print("\nStep 7: Final Visualization & Profiling...")
    final_model = KMeans(n_clusters=OPTIMAL_K_DS2, init='k-means++', n_init=10, random_state=42)
    labels = final_model.fit_predict(X_features_ds2)
    df_cleaned_ds2['segment_id'] = labels

    # --- SEGMENT PROFILING & NAMING (DS2) ---
    print("\n--- Segment Profiles (Dataset 2) ---")
    print("\n[Numerical Profile (Mean Values)]")
    numerical_profile_ds2 = df_cleaned_ds2.groupby('segment_id')[numerical_features_ds2].mean()
    print(numerical_profile_ds2.round(2))
    wandb.log({"Final/Numerical Profile": wandb.Table(dataframe=numerical_profile_ds2.reset_index())})

    print("\n[Categorical Profile (Mode)]")
    categorical_profile_ds2 = df_cleaned_ds2.groupby('segment_id')[categorical_features_ds2].agg(lambda x: x.mode().iloc[0])
    print(categorical_profile_ds2)
    wandb.log({"Final/Categorical Profile": wandb.Table(dataframe=categorical_profile_ds2.reset_index())})

    # --- DYNAMIC SEGMENT NAMING LOGIC (DS2) - IMPROVED ---
    # 1. Get global averages to compare against
    avg_tox = df_cleaned_ds2['toxicity_score'].mean()
    avg_sent = df_cleaned_ds2['sentiment_score'].mean()
    avg_eng = df_cleaned_ds2['avg_engagement_rate'].mean()

    def name_segment_ds2(row):
        name = []
        
        # Priority 1: Toxicity (as it's a key behavioral flag)
        if row['toxicity_score'] > (avg_tox * 1.5): # 50% above avg
            name.append(f"Toxic ({row['topic_category']})")
        elif row['toxicity_score'] < (avg_tox * 0.5): # 50% below avg
            name.append(f"Wholesome ({row['topic_category']})")
        else:
            name.append(f"General ({row['topic_category']})")

        # Priority 2: Engagement Level
        if row['avg_engagement_rate'] > (avg_eng * 1.5): # 50% above avg
            name.append("Super-Engagers")
        elif row['avg_engagement_rate'] > (avg_eng * 1.1): # 10% above avg
            name.append("High-Engagers")
        elif row['avg_engagement_rate'] < (avg_eng * 0.7): # 30% below avg
            name.append("Low-Engagers")
        else:
            name.append("Average-Engagers")

        # Priority 3: Sentiment (if not toxic)
        if 'Toxic' not in name[0]:
            if row['sentiment_score'] > (avg_sent + 0.1):
                name.append(f"({row['emotion_type']})") # e.g., (Joy)
            elif row['sentiment_score'] < (avg_sent - 0.1):
                name.append(f"({row['emotion_type']})") # e.g., (Sadness)
        
        # Priority 4: Platform
        name.append(f"on {row['platform']}")

        return " ".join(name)
    # --- END LOGIC ---
    
    profile_summary_ds2 = numerical_profile_ds2.join(categorical_profile_ds2, how="left")
    segment_name_map_ds2 = {idx: name_segment_ds2(row) for idx, row in profile_summary_ds2.iterrows()}

    print("\n--- Meaningful Segment Definitions (DS2) ---")
    print("Significance: These names describe the 'persona' of each cluster.")
    for seg_id, name in segment_name_map_ds2.items():
        print(f"Segment {seg_id}: {name}")
        
    df_cleaned_ds2['segment'] = df_cleaned_ds2['segment_id'].map(segment_name_map_ds2)
    # --- END PROFILING ---
    
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_features_ds2)
    
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df_cleaned_ds2['segment'], alpha=0.6)
    plt.title('User Segments (PCA) - Dataset 2')
    plt.savefig('ds2_pca.png') 
    wandb.log({"Final/PCA_Plot": wandb.Image('ds2_pca.png')}) 
    plt.show()
    
    df_cleaned_ds2.to_csv('Dataset2_Segmented.csv', index=False)
    print(">> Dataset 2 saved as 'Dataset2_Segmented.csv'")

except Exception as e:
    print(f"\n!!! Error in Dataset 2: {e} !!!")
finally:
    if ds2_run:
        wandb.finish() 


# --- DATASET 3: Social Media Users ---

print("\n\n" + "="*80)
print("STARTING DATASET 3: Social Media Users")
print("="*80)

ds3_run = wandb.init(
    project="social-media-segmentation",
    name="Dataset 3 - Users",
    reinit=True
)

try:
    print("Step 1 & 2: Loading and preparing data...")
    # Make sure to update these paths to your local file locations
    df_ds3 = pd.read_csv('D:\\Project\\Dataset 3\\SocialMediaUsersDataset.csv')
    df_ds3['Age'] = 2024 - pd.to_datetime(df_ds3['DOB'], errors='coerce').dt.year
    df_ds3['Interests'] = df_ds3['Interests'].astype(str).str.replace("'", "").str.replace('"', '')
    print("Data prepared.")

    print("\nStep 3: Analyzing & Cleaning...")
    # Impute missing Age *before* outlier detection
    age_mean = df_ds3['Age'].mean()
    print(f"Imputing {df_ds3['Age'].isna().sum()} missing ages with mean value ({age_mean:.1f})")
    df_ds3['Age'] = df_ds3['Age'].fillna(age_mean)
    
    plt.figure(figsize=(10, 3))
    sns.boxplot(x=df_ds3['Age'], orient='h')
    plt.title('Boxplot for User Age (Dataset 3, After Imputing)')
    plt.savefig('ds3_boxplot.png') 
    wandb.log({"Data Analysis/Boxplot (Before)": wandb.Image('ds3_boxplot.png')}) 
    plt.show()
    
    Q1 = df_ds3['Age'].quantile(0.25)
    Q3 = df_ds3['Age'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound_ds3 = Q1 - 1.5 * IQR
    upper_bound_ds3 = Q3 + 1.5 * IQR
    df_cleaned_ds3 = df_ds3[(df_ds3['Age'] >= lower_bound_ds3) & (df_ds3['Age'] <= upper_bound_ds3)].copy()
    rows_removed = len(df_ds3) - len(df_cleaned_ds3)
    print(f"Removed {rows_removed} outliers based on Age.")
    wandb.log({"Data Analysis/Rows Removed": rows_removed}) 
    
    if df_cleaned_ds3.empty:
        raise Exception("No data remaining after outlier removal. Stopping execution for Dataset 3.")

    print(f"\n--- Data Preview (DS3 After Trimming) ---")
    print(df_cleaned_ds3.head())
    print("\nCleaned Data Description (DS3):")
    print(df_cleaned_ds3.describe().round(2))
    
    stats_ds3 = {
        "mean": df_cleaned_ds3['Age'].mean(),
        "trimmed_mean_10": trim_mean(df_cleaned_ds3['Age'], 0.1)
    }
    print(f"[Stats 'Age'] Mean: {stats_ds3['mean']:.2f}")
    print(f"Trimmed Mean (10%): {stats_ds3['trimmed_mean_10']:.2f}")
    wandb.log({"Data Analysis/Stats": stats_ds3}) 

    print("\nStep 4: ML Pipelines...")
    numerical_features_ds3 = ['Age']
    categorical_features_ds3 = ['Gender']
    text_features_ds3 = 'Interests'

    preprocessor_ds3 = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_features_ds3),
            ('cat', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), categorical_features_ds3),
            ('text', TfidfVectorizer(max_features=50, stop_words='english'), text_features_ds3)
        ], remainder='drop')

    X_features_ds3 = preprocessor_ds3.fit_transform(df_cleaned_ds3)
    if hasattr(X_features_ds3, 'toarray'):
        X_features_ds3 = X_features_ds3.toarray()
    print(f"Features created. Shape: {X_features_ds3.shape}")

    # Determine sample size, ensuring it's not larger than the dataset
    n_samples_total = X_features_ds3.shape[0]
    n_sample_size = min(10000, n_samples_total)
    
    print(f"\nTaking a {n_sample_size} sample for faster experiments...")
    np.random.seed(42)
    sample_idx = np.random.choice(n_samples_total, n_sample_size, replace=False)
    X_sample_ds3 = X_features_ds3[sample_idx]
    wandb.config.update({"sample_size": n_sample_size}) 

    print("\nStep 5: Finding optimal K (on sample)...")
    wcss = []
    max_k_ds3 = min(10, n_sample_size - 1)
    if max_k_ds3 < 2:
        raise Exception(f"Not enough samples ({n_sample_size}) to run KMeans.")
    k_range_ds3 = range(2, max_k_ds3 + 1)
    print(f"Running Elbow method for K from 2 to {max_k_ds3}")
    
    for k in k_range_ds3:
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
        kmeans.fit(X_sample_ds3)
        wcss.append(kmeans.inertia_)
        wandb.log({"Elbow/WCSS_sample": kmeans.inertia_, "Elbow/K": k}) 
    
    plt.figure(figsize=(8, 4))
    plt.plot(k_range_ds3, wcss, 'o--')
    plt.title('Elbow Method (Dataset 3 Sample)')
    plt.savefig('ds3_elbow.png') 
    wandb.log({"Elbow/Elbow Plot": wandb.Image('ds3_elbow.png')}) 
    plt.show()
    
    OPTIMAL_K_DS3 = 5 if max_k_ds3 >= 5 else max_k_ds3
    print(f"Chosen K for Dataset 3: {OPTIMAL_K_DS3}")
    wandb.config.update({"optimal_k": OPTIMAL_K_DS3}) 

    run_experiments(X_sample_ds3, OPTIMAL_K_DS3, "Dataset 3 (Sample)")
    run_kmeans_cv(X_sample_ds3, OPTIMAL_K_DS3, n_splits=5, dataset_name="Dataset 3 (Sample)")
    run_dbscan_analysis(X_sample_ds3, eps=1.0, min_samples=5, dataset_name="Dataset 3 (Sample)")

    print("\nStep 7: Final Visualization & Profiling (on full data)...")
    final_model = KMeans(n_clusters=OPTIMAL_K_DS3, init='k-means++', n_init=10, random_state=42)
    final_model.fit(X_features_ds3)
    df_cleaned_ds3['segment_id'] = final_model.predict(X_features_ds3)

    # --- SEGMENT PROFILING & NAMING (DS3) ---
    print("\n--- Segment Profiles (Dataset 3) ---")
    print("\n[Numerical Profile (Mean Values)]")
    numerical_profile_ds3 = df_cleaned_ds3.groupby('segment_id')[numerical_features_ds3].mean()
    print(numerical_profile_ds3.round(2))
    wandb.log({"Final/Numerical Profile": wandb.Table(dataframe=numerical_profile_ds3.reset_index())})
    
    print("\n[Categorical Profile (Mode & Top Interests)]")
    categorical_profile_ds3 = df_cleaned_ds3.groupby('segment_id')[categorical_features_ds3].agg(lambda x: x.mode().iloc[0])
    try:
        # Access the vectorizer directly
        tfidf_vectorizer_ds3 = preprocessor_ds3.named_transformers_['text']
        text_feature_names_ds3 = tfidf_vectorizer_ds3.get_feature_names_out()
        n_text_features_ds3 = len(text_feature_names_ds3)
        
        # Slice only the last 'n_text_features_ds3' from the cluster centers
        cluster_centers_tfidf_ds3 = final_model.cluster_centers_[:, -n_text_features_ds3:]
        
        top_interests_ds3 = []
        for i in range(OPTIMAL_K_DS3):
            top_indices_ds3 = cluster_centers_tfidf_ds3[i].argsort()[-3:][::-1]
            top_interests_ds3.append(', '.join([text_feature_names_ds3[idx] for idx in top_indices_ds3]))
        categorical_profile_ds3['Top_Interests'] = top_interests_ds3
    except Exception as e:
        print(f"Could not get interests for profiling: {e}")
    print(categorical_profile_ds3)
    wandb.log({"Final/Categorical Profile": wandb.Table(dataframe=categorical_profile_ds3.reset_index())})
    
    # --- DYNAMIC SEGMENT NAMING LOGIC (DS3) ---
    # --- NEW LOGIC BASED ON USER INSIGHT ---
    
    def name_segment_ds3(row):
        # 1. Get Gender (the 'mode' from categorical profile)
        gender = str(row['Gender']) # e.g., "Female", "Male"
        
        # 2. Get Age (the 'mean' from numerical profile)
        age = int(row['Age'])
        
        # 3. Determine Age Bracket based on your examples
        if age < 35:
            age_bracket = "Young"
        elif age < 50:
            age_bracket = "Mid-Life"
        else:
            age_bracket = "Senior"
            
        # 4. Combine them into a clear, demographic-first name
        # Format: "Young Females (Avg Age: 28)"
        return f"{age_bracket} {gender}s (Avg Age: {age})"
    # --- END LOGIC ---

    profile_summary_ds3 = numerical_profile_ds3.join(categorical_profile_ds3, how="left")
    segment_name_map_ds3 = {idx: name_segment_ds3(row) for idx, row in profile_summary_ds3.iterrows()}
    
    print("\n--- Meaningful Segment Definitions (DS3) ---")
    print("Significance: These names describe the 'persona' of each cluster.")
    for seg_id, name in segment_name_map_ds3.items():
        print(f"Segment {seg_id}: {name}")
        
    df_cleaned_ds3['segment'] = df_cleaned_ds3['segment_id'].map(segment_name_map_ds3)
    # --- END PROFILING ---

    # Plot PCA using the sample data for performance
    print("\nStep 7a: Standard Visualization (PCA)...")
    labels_for_sample = final_model.predict(X_sample_ds3) 
    pca = PCA(n_components=2, random_state=42)
    X_pca = pca.fit_transform(X_sample_ds3)
    
    # Map segment IDs from sample back to names for plotting
    sample_segments_named = pd.Series(labels_for_sample).map(segment_name_map_ds3)
    
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=sample_segments_named, palette='viridis', alpha=0.6)
    plt.title('User Segments (PCA) - Dataset 3 Sample')
    plt.xlabel('Principal Component 1'); plt.ylabel('Principal Component 2')
    plt.savefig('ds3_pca.png') 
    wandb.log({"Final/PCA_Plot": wandb.Image('ds3_pca.png')}) 
    plt.show()
    
    
    # --- NEW VISUALIZATION: t-SNE ---
    print("\nStep 7b: Alternate Visualization (t-SNE)...")
    print("Running t-SNE on sample data (this may take a minute)...")
    
    # We already have 'labels_for_sample' and 'sample_segments_named' from above
    
    # Initialize t-SNE
    # Perplexity is ~ number of neighbors, 30-50 is a good start
    # init='pca' helps stabilize the layout
    tsne = TSNE(n_components=2, perplexity=30, random_state=42, n_iter=300, init='pca')
    
    # Fit and transform the *sample* data
    X_tsne = tsne.fit_transform(X_sample_ds3)
    
    print("t-SNE complete. Plotting...")

    # Plot the t-SNE results
    plt.figure(figsize=(10, 7))
    sns.scatterplot(x=X_tsne[:,0], y=X_tsne[:,1], hue=sample_segments_named, palette='viridis', alpha=0.7)
    plt.title('User Segments (t-SNE) - Dataset 3 Sample')
    plt.xlabel('t-SNE Component 1'); plt.ylabel('t-SNE Component 2')
    plt.savefig('ds3_tsne.png') 
    wandb.log({"Final/t-SNE_Plot": wandb.Image('ds3_tsne.png')}) 
    plt.show()
    # --- END NEW VISUALIZATION ---

    
    df_cleaned_ds3.to_csv('Dataset3_Segmented.csv', index=False)
    print(">> Dataset 3 saved as 'Dataset3_Segmented.csv'")

except Exception as e:
    print(f"\n!!! Error in Dataset 3: {e} !!!")
finally:
    if ds3_run:
        wandb.finish() 


# --- DATASET 4: Instagram Visits ---

print("\n\n" + "="*80)
print("STARTING DATASET 4: Instagram Visits")
print("="*80)

ds4_run = wandb.init(
    project="social-media-segmentation",
    name="Dataset 4 - Insta Visits",
    reinit=True
)

try:
    print("Step 1: Loading data...")
    # Make sure to update these paths to your local file locations
    df_ds4 = pd.read_csv('D:\\Project\\Dataset 4\\Instagram visits clustering.csv')
    print("Data loaded.")

    print("\nStep 2/3: Analyzing...")
    print(df_ds4.head())
    print("\nData Description (DS4):")
    stats_ds4 = df_ds4.describe().round(2)
    print(stats_ds4)
    wandb.log({"Data Analysis/Stats": stats_ds4.to_dict()}) 
    
    X_features_ds4 = df_ds4[['Instagram visit score', 'Spending_rank(0 to 100)']].values
    print(f"Features ready. Shape: {X_features_ds4.shape}")

    print("\nStep 5: Finding optimal K...")
    wcss = []
    max_k_ds4 = min(10, X_features_ds4.shape[0] - 1)
    if max_k_ds4 < 2:
        raise Exception(f"Not enough samples ({X_features_ds4.shape[0]}) to run KMeans.")
    k_range_ds4 = range(2, max_k_ds4 + 1)
    print(f"Running Elbow method for K from 2 to {max_k_ds4}")

    for k in k_range_ds4:
        kmeans = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=42)
        kmeans.fit(X_features_ds4)
        wcss.append(kmeans.inertia_)
        wandb.log({"Elbow/WCSS": kmeans.inertia_, "Elbow/K": k}) 
    
    plt.figure(figsize=(8, 4))
    plt.plot(k_range_ds4, wcss, 'o--')
    plt.title('Elbow Method (Dataset 4)')
    plt.savefig('ds4_elbow.png') 
    wandb.log({"Elbow/Elbow Plot": wandb.Image('ds4_elbow.png')}) 
    plt.show()
    
    OPTIMAL_K_DS4 = 4 if max_k_ds4 >= 4 else max_k_ds4
    print(f"Chosen K for Dataset 4: {OPTIMAL_K_DS4}")
    wandb.config.update({"optimal_k": OPTIMAL_K_DS4}) 

    run_experiments(X_features_ds4, OPTIMAL_K_DS4, "Dataset 4")
    run_kmeans_cv(X_features_ds4, OPTIMAL_K_DS4, n_splits=5, dataset_name="Dataset 4")
    run_dbscan_analysis(X_features_ds4, eps=5, min_samples=5, dataset_name="Dataset 4")

    print("\nStep 7: Final Visualization & Profiling...")
    # This dataset is already 2D, so no PCA/t-SNE is needed.
    final_model = KMeans(n_clusters=OPTIMAL_K_DS4, init='k-means++', n_init=10, random_state=42)
    labels = final_model.fit_predict(X_features_ds4)
    df_ds4['segment_id'] = labels
    
    # --- SEGMENT PROFILING & NAMING (DS4) ---
    print("\n--- Segment Profiles (Dataset 4) ---")
    print("\n[Numerical Profile (Mean Values)]")
    numerical_profile_ds4 = df_ds4.groupby('segment_id')[['Instagram visit score', 'Spending_rank(0 to 100)']].mean()
    print(numerical_profile_ds4.round(2))
    wandb.log({"Final/Numerical Profile": wandb.Table(dataframe=numerical_profile_ds4.reset_index())})
    
    avg_visit = df_ds4['Instagram visit score'].mean()
    avg_spend = df_ds4['Spending_rank(0 to 100)'].mean()
    
    def name_segment_ds4(row):
        if row['Instagram visit score'] > avg_visit and row['Spending_rank(0 to 100)'] > avg_spend:
            return "High-Value (High Visit, High Spend)"
        elif row['Instagram visit score'] > avg_visit and row['Spending_rank(0 to 100)'] <= avg_spend:
            return "Window Shoppers (High Visit, Low Spend)"
        elif row['Instagram visit score'] <= avg_visit and row['Spending_rank(0 to 100)'] > avg_spend:
            return "Potential Customers (Low Visit, High Spend)"
        else:
            return "Low-Activity Users (Low Visit, Low Spend)"

    segment_name_map_ds4 = {idx: name_segment_ds4(row) for idx, row in numerical_profile_ds4.iterrows()}

    print("\n--- Meaningful Segment Definitions (DS4) ---")
    print("Significance: These names describe the 'persona' of each cluster.")
    for seg_id, name in segment_name_map_ds4.items():
        print(f"Segment {seg_id}: {name}")
        
    df_ds4['segment'] = df_ds4['segment_id'].map(segment_name_map_ds4)
    # --- END PROFILING ---

    df_ds4.to_csv('Dataset4_Segmented.csv', index=False)
    print(">> Dataset 4 saved as 'Dataset4_Segmented.csv'")

    plt.figure(figsize=(10, 7))
    sns.scatterplot(data=df_ds4, x='Instagram visit score', y='Spending_rank(0 to 100)', hue='segment', s=100)
    
    plt.scatter(final_model.cluster_centers_[:, 0], final_model.cluster_centers_[:, 1], 
                s=200, c='yellow', edgecolors='black', label='Centroids')
    plt.title('Final Segments - Dataset 4')
    plt.xlabel('Visit Score'); plt.ylabel('Spending Rank')
    plt.legend()
    plt.savefig('ds4_final_plot.png') 
    wandb.log({"Final/Final_Plot": wandb.Image('ds4_final_plot.png')}) 
    plt.show()

except Exception as e:
    print(f"\n!!! Error in Dataset 4: {e} !!!")
finally:
    if ds4_run:
        wandb.finish() 


print("\n\n" + "="*80)
print("ALL DATASET PROCESSING COMPLETE")
print("="*80)
```
